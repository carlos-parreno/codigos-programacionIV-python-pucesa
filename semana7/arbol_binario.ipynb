```python
# importamos matplotlib para graficar
import matplotlib.pyplot as plt

# importamos networkx (grafos dirigidos o no dirigidos)
import networkx as nx

# creamos un grafo dirigido
G = nx.DiGraph()

# definir aristas y sus relaciones padre-hijo
edges = [
    (1, 2),
    (1, 3),
    (2, 4),
    (2, 10),
    (10, 11),
]

# agregamos las aristas al grafo, y se crean automaticamente los nodos
G.add_edges_from(edges)

# definimos posiciones en el plano
pos = {
    1: (0, 2),
    2: (-1.5, 1),
    3: (1.5, 1),
    4: (-3, 0),
    10: (-1.2, 0),
    11: (-1.4, -1)
}

# creamos una nueva figura
plt.figure(figsize = (10, 6))

# dibujar grafo

nx.draw(
    G, pos,
    with_labels = True,
    node_color = "lightblue",
    edge_color = "red",
    node_size = 1500,
    font_size = 12,
    font_weight = "bold",
    arrows = False 
)

#titulo
plt.title("Arbol Binario de Busqueda: [1, 2, 3, 4, 10, 11]", fontsize = 20)

#ocultar los ejes
plt.axis("off")

# mostrar grafico
plt.show()
```

```python
# importamos la libreria para graficar
import matplotlib.pyplot as plt

# importamos libreria para hacer grafos dirigidos o no-dirigidos
import networkx as nx

# creacion de grafo
H = nx.DiGraph()

# creacion de relaciones (aristas) y sus relaciones padre-hijo
edges = [
    ('/', '+'),
    ('/', '-'),
    ('+', 22),
    ('+', 3),
    ('-', 5),
    ('-', 59)
]

# añade las aristas al grafo H y crea los nodos
H.add_edges_from(edges)

#definimos posiciones (x, y) en el plano para cada nodo
pos = {
    '/': (0, 2),
    '+': (-1.5, 1),
    '-': (1.5, 1),
    22: (-2.2, 0),
    3: (-0.8, 0),
    5: (0.8, 0),
    59: (2.2, 0)
}

plt.figure(figsize = (8, 5))

nx.draw(
    H, pos,
    with_labels = True,
    node_color = 'red',
    node_size = 2000,
    font_size = 14,
    font_weight = 'bold',
    arrows = True    
)

plt.title("Árbol de expresión (22+3)/(5-59)")

plt.show()
```
